# Native Logic Gates

In quantum computers, there are native (physical) operations that the hardware implements directly. This depends on the architecture. IBM: Rz, Sx, and CNOT (2 bits).

```python
!pip install qiskit-ibm-runtime -q
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# 1 qubit

qc = QuantumCircuit(1)
qc.h(0)

print("Original circuit")
display(qc.draw("mpl"))

# IBM default - force to use native gates
basic_gates = ["rz", "sx", "x", "cx"]
backend = AerSimulator()

tqc = transpile(
    qc,
    backend=backend,
    basis_gates=basic_gates,
    optimization_level = 0
)

print("transpiled circuit")
display(tqc.draw("mpl"))
```
