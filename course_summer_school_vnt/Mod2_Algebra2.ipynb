```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo um Ket: |ψ> e |φ>
ket_psi = np.array([[1], [0]])  # Representa |ψ> no espaço vetorial
ket_phi = np.array([[0], [1]])  # Representa |φ> no espaço vetorial
```

```python
# Exibindo com colchetes
print("|ψ> = ")
print(ket_psi)
```

```python
# Exibindo com colchetes
print("|φ> = ")
print(ket_phi)
```

```python
# Definindo um Bra: <ψ| e <φ|
bra_psi = ket_psi.conj().T  # <ψ|, transposta conjugada de |ψ>
bra_phi = ket_phi.conj().T  # <φ|, transposta conjugada de |φ>
```

```python
# Exibindo com colchetes
print("<ψ| = ")
print(bra_psi)
```

```python
# Exibindo com colchetes
print("<φ| = ")
print(bra_phi)
```

```python
# Produto Interno (Bra-Ket): <ψ|φ>
inner_product = np.dot(bra_psi, ket_phi)
print("Produto Interno <ψ|φ>:", inner_product)
```

```python
# Produto Externo (Ket-Bra): |ψ><φ|
outer_product = np.dot(ket_psi, bra_phi)
print("Produto Externo |ψ><φ|:\n", outer_product)
```

```python
# Produto Tensorial (Produto de Kronecker) entre |ψ> e |φ>
tensor_product = np.kron(ket_psi, ket_phi)
print("Produto Tensorial |ψ>|φ>:\n", tensor_product)
```
