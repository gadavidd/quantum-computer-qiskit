```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando vetores
vetor1 = np.array([1, 2, 3])
vetor2 = np.array([4, 5, 6])

# Exibindo com colchetes
print("Vetor1:")
print(vetor1)
print("Vetor2:")
print(vetor2)
```

```python
vetor3 = np.array([[1], [2], [3]])
vetor4 = np.array([[4], [5], [6]])
# Exibindo com colchetes
print("Vetor3:")
print(vetor3)
print("Vetor4:")
print(vetor4)
```

```python
# Criando matrizes
matriz1 = np.array([[1, 2], [3, 4]])
matriz2 = np.array([[5, 6], [7, 8]])

print("Matriz1:")
print(matriz1)
print("Matriz2:")
print(matriz2)
```

```python
# Adição de matrizes
matriz_soma = matriz1 + matriz2
print("Adição de Matrizes:\n", matriz_soma)
```

```python
# Subtração de matrizes
matriz_subtracao = matriz1 - matriz2
print("Subtração de Matrizes:\n", matriz_subtracao)
```

```python
# Multiplicação de matrizes
matriz_multiplicacao = np.dot(matriz1, matriz2)
print("Multiplicação de Matrizes:\n", matriz_multiplicacao)
```

```python
# Matriz identidade
matriz_identidade = np.eye(2)
print("Matriz Identidade:\n", matriz_identidade)
```

```python
# Matriz transposta
matriz_transposta = np.transpose(matriz1)
print("Matriz Transposta:\n", matriz_transposta)
```

```python
# Números complexos
complexo1 = np.array([[1 + 2j, 3 + 4j], [5 + 6j, 7 + 8j]])
complexo2 = np.array([[1 - 2j, 3 - 4j], [5 - 6j, 7 - 8j]])
```

```python
print("complexo1:")
print(complexo1)
```

```python
# Matriz transposta conjugada (também conhecida como adjunta ou hermitiana)
matriz_conjugada_transposta = np.conjugate(np.transpose(complexo1))
print("Matriz Transposta Conjugada:\n", matriz_conjugada_transposta)
```

```python
# Traço da matriz (soma dos elementos da diagonal principal)
traco_matriz = np.trace(matriz1)
print("Traço da Matriz:", traco_matriz)
```
