**Estados de Bell**

Aqui vamos mostrar como é construido um doos estados de Bell através do Qiskit. Para isso, vamos considerar um sistema de dois qubits. Como explicado nos slides, o estado de Bell é alcançado com a utilização das portas quânticas Hadamard e CNOT. Mas antes, é necessário construir um circuito com 2 qubits e 2 bits clássicos registradores.

```python
# Importe as bibliotecas necessárias para executar o projeto.
from qiskit import QuantumCircuit, transpile
from qiskit import QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import qiskit.quantum_info as qi
```

```python
# Cria os registros para a criação dos qubits e bits
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
# Cria o circuito e faz o emaranhamento gerando um dos estados de Bell
```

```python
circuit = QuantumCircuit(qr, cr)
circuit.h(qr[0])
circuit.cx(qr[0], qr[1])
circuit.measure(qr, cr)
circuit.draw('mpl')
```

Em seguida, é feita a simulação do circuito na qual retorna as probabilidades dos possíveis estados, |00> e |11>.
Note que a projeção dos estados de Bell retorna uma probabilidade
próxima à 50% para ambos os estados |00> e |11>. Se executarmos essas linhas de códigos inúmeras vezes, o resultado sempre será 50%.

```python
# Use o simulador Qiskit Aer para executar o circuito
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
result = simulator.run(compiled_circuit).result()
```

```python
# Obtenha a contagem dos resultados
counts = result.get_counts()
print("Counts:", counts)
# Visualize os resultados
plot_histogram(counts)
```

É importante dizer que além dos estados de Bell, existem outros tipos de estados emaranhados.  Mas no caso dos estados de Bell, eles são um exemplo clássico de estados maximamente emaranhados para dois qubits. Estes estados têm a propriedade de maximizar o emaranhamento entre dois qubits, o que significa que o conhecimento do estado de um qubit fornece total informação sobre o estado do outro, independentemente da distância entre eles.
