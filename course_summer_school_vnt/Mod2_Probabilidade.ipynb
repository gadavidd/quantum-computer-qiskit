#Simulando o lançamento de um dado:

Vamos criar um código em Python que simula um cenário interessante usando probabilidade. Vou considerar um exemplo de lançamento de um dado de 6 lados. Vamos simular 100 lançamentos e analisar os resultados.

**Esse exemplo envolve os seguintes passos:**

Gerar dados aleatórios simulando os lançamentos do dado.

Calcular a média e mediana dos resultados.

Calcular a variância e o desvio padrão.

Gerar um histograma para visualizar a distribuição dos resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Configurações
np.random.seed(42)  # Para reprodutibilidade -  garante que os números aleatórios gerados serão os mesmos em cada execução do programa!
num_lancamentos = 100
dados = np.random.randint(1, 7, num_lancamentos)  # Gera valores entre 1 e 6
```

```python
# Cálculo de estatísticas
media = np.mean(dados)
mediana = np.median(dados)
variancia = np.var(dados)
desvio_padrao = np.std(dados)
```

```python
# Exibição das estatísticas
print(f"Média: {media}")
print(f"Mediana: {mediana}")
#print(f"Moda: {moda}")
print(f"Variância: {variancia}")
print(f"Desvio Padrão: {desvio_padrao}")
```

```python
# Plot do histograma
plt.figure(figsize=(10, 6))
sns.histplot(dados, bins=6, kde=True, color='skyblue', edgecolor='black')
plt.title("Distribuição dos Resultados dos Lançamentos do Dado")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
```
